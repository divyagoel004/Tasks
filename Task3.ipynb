```python
from langchain.tools import Tool
from langchain.agents import initialize_agent
from langchain.agents import AgentType
from together import Together
from langchain_openai import ChatOpenAI
import requests
from bs4 import BeautifulSoup
import os

os.environ['TOGETHER_API_KEY'] = "666e8dc46b45912071558831227d603654035a8f70c3e9f0a41f6f0f9661ad98"
```

```python
url='https://animemangatoon.com/castle-swimmer-unveiling-new-prophecy/'
def scrape_website(url: str) -> str:
    url = url.replace('"', '')
    return BeautifulSoup(requests.get(url).content, 'html.parser').get_text()[:1000]

model = ChatOpenAI(
    base_url="https://api.together.xyz/v1",
    api_key=os.environ["TOGETHER_API_KEY"],
    model="mistralai/Mixtral-8x7B-Instruct-v0.1",
)
```

```python
tool = Tool(
    name="WebScraper",
    func=scrape_website,
    description="Fetch content from a website URL"
)
```

```python
agent=initialize_agent([tool],model,agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,verbose=True)
```

```python
agent.run("What is Castle Swimmer about?")
```
